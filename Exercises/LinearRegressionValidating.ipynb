```python
import math
import warnings

from IPython.display import display
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model
import statsmodels.formula.api as smf

# Display preferences.
%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format

# Suppress annoying harmless error.
warnings.filterwarnings(
    action="ignore",
    module="scipy",
    message="^internal gelsd"
)

data = pd.read_csv('C://duh/Thinkful/table_8_offenses_known_to_law_enforcement_utah_by_city_2013.csv')[:82]
data['Property crime'] = data['Property crime'].str.replace(',', '')
data['Property crime'] = pd.to_numeric(data['Property crime'])
data['Population'] = data['Population'].str.replace(',', '')
data['Population'] = pd.to_numeric(data['Population'])
data['Violent crime'] = data['Violent crime'].str.replace(',', '')
data['Violent crime'] = pd.to_numeric(data['Violent crime'])
data['Burglary'] = data['Burglary'].str.replace(',', '')
data['Burglary'] = pd.to_numeric(data['Burglary'])
data['Larceny-theft'] = data['Larceny-theft'].str.replace(',', '')
data['Larceny-theft'] = pd.to_numeric(data['Larceny-theft'])
data['Motor vehicle theft'] = data['Motor vehicle theft'].str.replace(',', '')
data['Motor vehicle theft'] = pd.to_numeric(data['Motor vehicle theft'])
pred_data = data[['Population', 'Violent crime', 'Murder and nonnegligent manslaughter', 
                  'Rape (revised definition)1', 'Rape (legacy definition)2', 'Robbery', 
                  'Aggravated assault', 'Burglary', 'Larceny-theft', 'Motor vehicle theft']].fillna(0)
target = data['Property crime'].fillna(0)
pred_data
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
from sklearn.model_selection import train_test_split
# Use train_test_split to create the necessary training and test groups
X_train, X_test, y_train, y_test = train_test_split(pred_data, target, test_size=0.2, random_state=20)
print('With 20% Holdout: ' + str(bnb.fit(X_train, y_train).score(X_test, y_test)))
print('Testing on Sample: ' + str(bnb.fit(pred_data, target).score(data, target)))
```
