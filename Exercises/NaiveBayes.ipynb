```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import BernoulliNB

data_raw = pd.read_csv('C://duh/Thinkful/UCIML/sentiment_labelled_sentences/amazon_cells_labelled.txt',
                       delimiter= '\t', header=None)
data_raw.columns = ['message', 'score']
keywords = ['great', 'good', 'love', 'pleased', 'excellent', 'happy', 'must have', 'best', 'impressed', 'recommend']
for key in keywords:
    data_raw[str(key)] = data_raw.message.str.contains(str(key), case=False)
data_raw['score'] = (data_raw['score'] == 1)
```

```python
sns.heatmap(data_raw.corr())
```

```python
data = data_raw[keywords]
target = data_raw['score']
```

```python
bnb = BernoulliNB()
bnb.fit(data, target)
y_pred = bnb.predict(data)
print("Number of mislabeled point out of a total {} points : {}".format(
    data.shape[0],
    (target != y_pred).sum()
))
```
