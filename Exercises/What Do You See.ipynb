# Direct Comparison of Countries by type
As stated before, the import and export comparison show the vise-versa nature of the countries. While the US maintains a generally consistant importation of oil, ranging from 6000-14000, the UK as continuously increased their importation. However, the UK maintains a consistant exportation of oil, while the US has continued to grow in exportation. US consumption of oil still greatly exceeds the UK.

```python
plt.figure(figsize=(30,10))
set_bin2 = [1000,2000,3000,4000,5000,6000,7000,8000,9000,10000,11000,12000,13000,14000,15000,16000]

plt.subplot(1, 2, 1)
plt.hist(year_importUS['Quantity'], bins=set_bin2, alpha=.5)
plt.hist(year_importUK['Quantity'], bins=set_bin2, alpha=.5)
plt.legend(('US (1000)', 'UK (1000)'), loc='best')
plt.title('US vs UK Import')
plt.xlabel('Metric ton, thousand')

plt.subplot(1, 2, 2)
plt.hist(year_exportUS['Quantity'], bins=set_bin2, alpha=.5)
plt.hist(year_exportUK['Quantity'], bins=set_bin2, alpha=.5)
plt.legend(('US (1000)', 'UK (1000)'), loc='best')
plt.title('US vs UK Export')
plt.xlabel('Metric ton, thousand')

plt.show()
```
